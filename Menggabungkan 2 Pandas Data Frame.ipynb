```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows =5
n_cols=5
#kolom 5x5
#dengan nama kolom adalah ABCDE
cols = tuple('ABCDE')

df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns = cols)

df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1,4]) #buang indesk 1 dan indeks 4
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0,3]) #buang indeks 0 dan indeks 3
df2
```

```python
#Menggabungjan 2 data frame dengan Inner 

df_inner = pd.merge(df1,df2, how='inner')
df_inner
```

```python
#Menggabungkan 2 data frame dengan Outer
df_outer = pd.merge(df1,df2, how='outer')
df_outer

#merupakan method dfmerge yang paling sering digunakan pada saat kerja
```
