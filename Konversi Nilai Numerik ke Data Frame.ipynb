```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 1
#akan memunculkan sebuah data frame dengan 10 baris dan 1 kolom
cols = ('usia',)

df=pd.DataFrame(np.random.randint(1,99, size=(n_rows,n_cols)), columns = cols)
#berfungsi untuk memunculkan angka random berupa integer antara 1-99

df
```

```python
df['kelompok_usia'] = pd.cut(df['usia'],
                             bins = [0,18,65,99],
                             labels = ['anak','dewasa','manula'])
df
#pd.cut berfungsi untuk mengelompokkan usia, lalu dengan pemotongan menggunakan bins dengan nilai integer sebagai pembatas,
#lalu parameters labels untuk kelompok usianya
```
